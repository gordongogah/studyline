```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
data = load_iris()
X = data['data']
y = data['target']
```

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import graphviz

tree = DecisionTreeClassifier(max_depth=3)
tree.fit(X,y)
graphviz.Source(export_graphviz(tree,
                               out_file=None,
                               feature_names=data['feature_names'],
                               class_names=data['target_names']
                               ))
```

```python
from sklearn.datasets import fetch_california_housing
from sklearn.tree import DecisionTreeRegressor
housing = fetch_california_housing()
X = housing['data']
y = housing['target']
```

```python
model = DecisionTreeRegressor(max_depth=3)
model.fit(X,y)
graphviz.Source(export_graphviz(model,
                               out_file=None,
                               feature_names=housing['feature_names']))
```

```python
from ipywidgets import IntSlider,interact
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score


scaler= StandardScaler()
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=1)
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)


def plot_features(max_depth=5,min_samples_leaf=2):
    tree = DecisionTreeClassifier(max_depth=max_depth,min_samples_leaf=min_samples_leaf)
    tree.fit(X_train,y_train)
    model_score = r2_score(y_test,tree.predict(X_test))
    
    x1,x2 = np.meshgrid(np.linspace(-2,2,100),np.linspace(-2,2,100))
    y_prob = tree.predict_proba(np.hstack([x1.reshape(-1,1),x2.reshape(-1,1)]))[:,1]
    plt.contourf(x1,x2,y_prob.reshape(50,50),16,cmap=plt.cm.bwr,alpha=0.75)
    plt.colorbar()
    
    plt.scatter(X[:,2],X[:,3],c=y,edgecolors='white',cmap=plt.cm.bwr)
    plt.xlabel('$_1_$')
    plt.ylabel('gogah')
    plt.title('accuracy_score {}'.format(model_score))
    
sample_slider = IntSlider(min=1,max=20,steps=1,description='min_sample_split')
depth_slider = IntSlider(min=1,max=40,steps=1,description='max_depth')
interact(plot_features,min_samples_leaf=sample_slider,max_depth=depth_slider)
    
```

```python
from sklearn.datasets import make_moons
X,y = make_moons(n_samples=250,noise=0.2,random_state=0)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)

def plot_features(max_depth=5,min_samples_leaf=2):
    tree = DecisionTreeClassifier(max_depth=max_depth,min_samples_leaf=min_samples_leaf)
    tree.fit(X_train,y_train)
    accuracy = tree.score(X_test,y_test)
    
    X1,X2 = np.meshgrid(np.linspace(-2,3),np.linspace(-2,2))
    y_prob = tree.predict_proba(np.hstack((X1.reshape(-1,1),X2.reshape(-1,1))))[:,1]
    plt.contourf(X1,X2,y_prob.reshape(50,50),16,cmap=plt.cm.bwr,alpha=0.75)
    plt.colorbar()
    
    plt.scatter(X[:,0],X[:,1],c=y,edgecolors='white',cmap=plt.cm.bwr)
    
depth = IntSlider(min=1,max=40,steps=1,description='depth')
sample_slider = IntSlider(min=1,max=20,steps=1,description='sample leaf')
interact(plot_features,max_depth=depth,max_samples_leaf=sample_slider)
```
